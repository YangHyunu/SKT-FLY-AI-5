```python
import numpy as np
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
```

# Matplotlib

```python
X = np.array([1,2,3,7,15,16])
X
y= np.sin(X)
plt.plot(X,y)
```

```python
x = np.arange(25)
y= np.sin(x)
plt.plot(x,y,c='blue',linestyle='--')
# 타이틀을 달고 싶을 때
plt.title('SIN GRAPH')

# X,y 축의 레이블을 달고 싶을 떄
plt.xlabel('X')
plt.ylabel('y')
plt.show()
```

```python
x = np.arange(0,25,0.1) # 0~25 까지 0.1단위로 생성
y= np.sin(x)
z= np.cos(x)
plt.plot(x,y)
plt.plot(x,z,c='orange') # 두 그래프가 겹쳐져서 그려짐 즉, 한 axes에 두 그래프가 겹쳐 그려진다.

# 타이틀을 달고 싶을 때
plt.title('SIN GRAPH')

# X,y 축의 레이블을 달고 싶을 떄
plt.xlabel('X')
plt.ylabel('y')
plt.figure(figsize=(10,5))
plt.plot(x,y,label='sin(x)')
plt.plot(x,z,c='orange',label='cos(x)')
plt.legend()
plt.show()

```

```python
# 요즘 방법 객체지향 방법
x = np.arange(0,25,0.1) # 0~25 까지 0.1단위로 생성
y= np.sin(x)
z= np.cos(x)
fig, ax = plt.subplots() # figure와 axis를 리턴함
fig.set_size_inches(4,4)
ax.plot(x,y,label='sin(x)')
ax.plot(x,z,c='orange',label='cos(x)')
ax.set_xlabel('X values')
ax.set_ylabel('Y values')
ax.set_title('GRAPH')
ax.legend()
plt.show()
```

## 여러개의 플롯 그리기

```python
x=np.linspace(0,10,30)
y=np.sin(x)
z=np.cos(x)
w=np.tan(x)
```

```python
plt.figure(figsize=(12,5))
plt.subplot(2,2,1)
plt.plot(x,y,'r-')
plt.subplot(2,2,2)
plt.plot(x,z,'g-')
plt.subplot(2,2,3)
plt.plot(x,w,'b-')
plt.subplot(2,2,4)
plt.plot(x,y,z,w)
plt.legend(['sin','cos','tan'])
plt.show()


```

```python
deck=[y,z,w,(y,z,w)]
for i in range(3):
    plt.subplot(2,2,i+1)
    plt.plot(x,deck[i])

plt.show()
```

```python
fig,axes= plt.subplots(2,2)
# axes[0][0].plot(x,y)
# axes[0][1].plot(x,z)
# axes[1][0].plot(x,w)
# axes[1][1].plot(x,y,z,w)
axes = axes.flatten() # axes의 자료형이 np.ndarray라 flatten할 수 있다. for 문으로 2차원 배열로 전개하기 까다로움으로..
for i in range(3):
  axes[i].plot(x,deck[i])
plt.show()
```

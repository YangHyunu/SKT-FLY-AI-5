```python
with open('/content/generative ai.txt') as f:
  text_gen_ai = f.read()
```

```python
from langchain.text_splitter import CharacterTextSplitter
# text_splitter 객체 생성
text_splitter = CharacterTextSplitter(
    separator = "\n\n",
    chunk_size = 1000,
    chunk_overlap  = 200,
    length_function = len,
)
```

```python
texts = text_splitter.split_text(text_gen_ai) # 각 문단의 글자 수가 1000이상으로 분할돼서 warning이 떳음
```

```python
print(texts[0])
print('-' * 100)
print(texts[1])
print('-' * 100)
print(texts[2])
```

```python
# 각 문단 단위에 몇 글자가 있는지 확인
char_list =[]
for i in range(len(texts)):
  char_list.append(len(texts[i]))

print(char_list) # 청크 사이즈를 1000으로 했는데 하나의 문단을 제외하고 다 초과했음 따라서 해당 방법은 잘 사용되지 않는다.
```

```python
print(type(texts)) # document 타입으로 바꾸기
docs = text_splitter.create_documents(texts) # document 객체로 변환
print(type(docs))
docs
```

```python
# 위에서 사용한 text_splitter는 청크사이즈가 넘어가는 문단을 무시한 상태로 split
# 따라서 실제로는 청크사이즈 이하가 될 때 까지 문단을 스플릿하는 RecursiveTextSplitter를 사용!
from langchain.text_splitter import RecursiveCharacterTextSplitter
# Recursive text_splitter 객체 생성
text_splitter = RecursiveCharacterTextSplitter(
    chunk_size = 1000,
    chunk_overlap  = 100,
    length_function = len,
)
```

```python
texts = text_splitter.create_documents([text_gen_ai])

print(texts[0].page_content)
print('-' * 100)
print(texts[1].page_content)
print('-' * 100)
print(texts[2].page_content)
```

```python
char_list = []
for i in range(len(texts)):
  char_list.append(len(texts[i].page_content))

print(char_list)
# 각 문단이 모두 청크사이즈 이하의 글자를 포함하도록 분할됐음 !
```

```python
from langchain.text_splitter import (
    Language,
    RecursiveCharacterTextSplitter,
)
```

```python
RecursiveCharacterTextSplitter.get_separators_for_language(Language.PYTHON) # 파이썬을 인식해서 소스코드를 분해할 수 있음
# class, def 단위로 소스코드를 분할 할 수 있다!
```

```python
python_code = '''
def hello_world():
  print("hello world")

# Call the function
hello_world()
  '''
```

```python
python_splitter = RecursiveCharacterTextSplitter.from_language(
    language=Language.PYTHON,
    chunk_size=50,
    chunk_overlap=0,
)
python_docs =python_splitter.create_documents([python_code])
python_docs
```

```python
import tiktoken

# Use get_encoding directly instead of encoding_for_model
tokenizer = tiktoken.get_encoding("cl100k_base")  # Directly get the tokenizer

def tiktoken_len(text):
    tokens = tokenizer.encode(text)
    return len(tokens) # 토큰의 수를 반환
```

```python
tiktoken_len('I love you')
```

```python
tiktoken_len('나는 너를 사랑해')
```

```python
text_splitter = RecursiveCharacterTextSplitter(
    chunk_size = 1000,
    chunk_overlap = 100,
    length_function = tiktoken_len,
)
```

```python
docs = text_splitter.split_documents(pages)
docs[0]
docs
```

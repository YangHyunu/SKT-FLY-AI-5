```python
from langchain.chat_models import AzureChatOpenAI # 대화형 모델
from langchain.schema import HumanMessage, SystemMessage, AIMessage
chatgpt= AzureChatOpenAI(
        deployment_name="dev-gpt-35-turbo",

)

message = [
    SystemMessage(content='You are a kind helpful assistant. that translates English to Korean'),
    HumanMessage(content='I love Korea')
]
response = chatgpt(message)
print(response)
```

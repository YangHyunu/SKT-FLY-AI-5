```python
# Web Document Loader HTML 이나 javascript 코드 만 걷어내고 웹페이지를 그대로 Document객체로 덤프하는 역할
# 원하는 내용만을 가져올 수 없어서 butifulsoup에 비해 잘 사용되지 않음

from langchain.document_loaders import WebBaseLoader
loader = WebBaseLoader('https://zdnet.co.kr/view/?no=20240807161811')
data = loader.load()

print(data)
print(data[0].page_content)
```

```python
# Pdf 를 읽어서 Document 객체로 반환
from langchain.document_loaders import PyPDFLoader

loader = PyPDFLoader('/content/DJI_Osmo_Pocket_3_User_Manual_v1.0_en.pdf')
data = loader.load()
print(data[0].page_content)
print(data[0]) # pdf를 참고한 gpt의 답변에서 metadata 에 해당 pdf 페이지를 알 수 있어서 유용하게 사용됨
```

```python
# [워드]docx 파일을 Document 형태로 파싱
from langchain.document_loaders import Docx2txtLoader

loader = Docx2txtLoader('/content/LLM이 만들어 내는 혁명과 교통산업.docx')
data = loader.load()
print(data)
print(data[0]) # 마지막 줄에 metadata가 출력되어 있어, 어떤 문서에서 가져왔는지 알 수 있다.
```

```python
# CSV파일을 읽어서 Document 형태로 파싱

from langchain.document_loaders import CSVLoader

loader = CSVLoader(file_path='/content/titanic3.csv',
                   csv_args= {
                       'delimiter': ',',
                       'quotechar': '"',
                       'fieldnames': ['PassengerId', 'Survived', 'Pclass', 'Name', 'Sex', 'Age', 'SibSp', 'Parch', 'Ticket', 'Fare', 'Cabin', 'Embarked']
                   }) #quotation charater를 "" 로 사용
data = loader.load()
print(data)
print(data[:10])
```
